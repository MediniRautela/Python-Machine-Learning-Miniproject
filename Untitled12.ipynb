```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
from sklearn import metrics
from sklearn import set_config
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/namesdataset.csv")
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.drop('Country_code',axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Country'].value_counts().plot(kind='pie')
```

```python
len(df['Country'].unique())
```

```python
df.groupby('Country')['Name'].size()
```

```python
# Model without using N-grams

train_data, test_data, train_labels, test_labels = train_test_split(
    df['Name'], df['Country'], test_size=0.2, random_state=42
)

# Create a Naive Bayes model
model = make_pipeline(CountVectorizer(), MultinomialNB())

# Training the model
model.fit(train_data, train_labels)

set_config(display='diagram')

# Making predictions on test set
predictions = model.predict(test_data)

# Evaluating accuracy
accuracy = metrics.accuracy_score(test_labels, predictions)
print(f"{accuracy:.2%}")

# Predicting nationalities using model
new_names = ['John Davis', 'Maria', 'Ahmed Khan']
print(model.predict(new_names))
```

```python
# Model using N-grams


# Spliting data into training and testing sets
train_data, test_data, train_labels, test_labels = train_test_split(
    df['Name'], df['Country'], test_size=0.2, random_state=42
)

# Create a Naive Bayes model with n-grams
ngram_model = make_pipeline(CountVectorizer(analyzer='char', ngram_range=(4, 6)), MultinomialNB())

# Training the model
ngram_model.fit(train_data, train_labels)

# Making predictions on test set
predictions = ngram_model.predict(test_data)

# Evaluating accuracy of the model
accuracy = metrics.accuracy_score(test_labels, predictions)
print(f"{accuracy:.2%}")

# pediction phase
new_names = ['Amy Johnson', 'Sunil Bisht', 'Sirac Liam']
print(ngram_model.predict(new_names))
```
